```python
#number of candies collected by 10 kind during a school event
#number of candies collected:[3,5,8,6,7,5,9,6,7].
#calculate the statistical mean , median , mode and standard deviation.
#visualize bar chart , histogram,normal distribution.
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Data
candies = [3, 5, 8, 6, 7, 5, 9, 6, 7,6]

# 1. Statistical Calculations
mean = np.mean(candies)
median = np.median(candies)
mode = max(set(candies),key=candies.count)
std_dev = np.std(candies)

# Print the results
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Standard Deviation: {std_dev}")

```

```python
import scipy.stats

# Sample dataset
candies = [3, 5, 8, 6, 7, 5, 9, 6, 7,6]

# mode ko find krenge 
mode_result = scipy.stats.mode(candies)

# phir print krenge Printing all modes
print("Modes:", mode_result.mode)
```

```python
#Visualization: Histogram and Normal Distribution
x = np.linspace(min(candies), max(candies), 1000)
pdf = norm.pdf(x, mean, std_dev)
#what is probability distribution and probability density function
# PD:i  i want to plot the probabilities of events on a graph and see how they are distributed.
# Discete variable                     Continuous variable (we cannot measure/count)
# whose outcomes can be counted         Height of people, temperature
# (decimal values are not considered)    []
# eg: dice, coins
# prob distrution , y axis:frequencies    prob density function
# Plot Histogram
plt.hist(candies, bins=5, alpha=0.6, color='skyblue', density=True, label='Histogram')

# Plot Normal Distribution Curve
plt.plot(x, pdf, color='red', label='Normal Distribution')


# Add Labels
plt.title('Candies Distribution')
plt.xlabel('Number of Candies')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
