##Part 3: Data Manipulation
#step 1: Renaming Columns Rename the IMDB_Rating column to Rating.
#step2: New Columns Add a new column called Decade that categorizes movies into decades based on their release year. For example, a movie released in 1995 belongs to the 1990s.
#step3: SortingSort the dataset by the Rating column in descending order and display the top 10 movies.

```python
import pandas as pa
df=pa.read_csv(r"C:\Users\chaud\OneDrive\Desktop\UNIT2PYTHONLIBRARIES\imdb_top_1000.csv")
df.head()
```

```python
#step1: rename the columns
df.rename(columns={'IMDB_Rating':'Rating'},inplace=True)
df.rename(columns={'Series_Title' : 'Beautiful_Title'},inplace=True)
print(df.columns)
```

```python
#step2: ab hum new columns ko add krenge dekhte hai kaise
df['Decade']=df['Released_Year'].str[:4] + '0s'
print(df[['Beautiful_Title', 'Released_Year' , 'Decade']].head()) # yaha actual me series_title col hai pr mene iska naam beautiful_title kr diya tha .
```

```python
#step 3 : hum ab sort krenge apne data ko
sorted_movies=df.sort_values('Rating', ascending=False)
print(sorted_movies.head(10))
```

##Part 4: Visualization (Optional)
#Simple Plots Create a bar plot showing the number of movies in each genre.
#Plot a histogram of the Rating column.

```python
#iske liye hume pehle ploting waali library import krni hogi
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
Genre_counts = df['Genre'].value_counts()
```

```python
# ab hum plot bnayenge dekhte hai kaise
plt.figure(figsize=(10,6))
Genre_counts.plot(kind='bar', color='skyblue', edgecolor='red')
plt.title("no of Genre_Movies: " ,fontsize=16)
plt.xlabel('Genre',fontsize=12)
plt.ylabel('Genre',fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.show()
```

```python
Genre_counts = df['Genre'].value_counts()
top_6_Genres=Genre_counts.head(6)
plt.figure(figsize=(10,6))
Genre_counts.plot(kind='pie', color=['green']*len('top_6_Genres'),startangle=140)
plt.title("no of Genre_Movies: " ,fontsize=16)
plt.ylabel('')
plt.show()
```
