##Histogram
#a histogram visualize  a numerical data to observe the count vs value in the data 

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.hist(df['No. of Clicks'])
```

```python
#ab humne kya kiya hai print toh pehle bhi kiya tha histogram chart pr ab humne ek acche and clear trike se histogram bnaya hai
#ab humne title diya hai iska 
#phir hume iska x-axis ka title diya hai
#phir humne iska y-axis ka title diya hai
#aur hume faltu ki chije nhi dikhayi de output me jaiase pehle dikh rahi thi isliye humne plt.show() ka use kiya hai
#method1
plt.hist(df['No. of Clicks'])
plt.title("Histogram to represent the clicks")
plt.xlabel("No of clicks")
plt.ylabel("Count of clicks")
plt.show()
```

```python
#method2,
sns.displot(df['No. of Clicks'],bins=10) #ye second trika hai plot bnane ka , bins mtlb hume data ko divide kiya hai
```

```python
#method3:
df['No. of Clicks'].plot(kind='hist')
```

```python
##Now move to the pie chart
#Piechart is a visualization which involves a a circle getting divided into multiple smaller slices to display the information
```

```python
#humne kya hiya hai browser column me jo bhi value hai unko count kiya hai
df['Browser'].value_counts()
```

```python
#isme humne kya kiya hai pie chart display kiya hai 
#aur yaha pr %,2f%% ka use humne chart me % dekhne ke liye kiya hai
#aur ye humare upr depend krega ki kitni decimal value chahiye if you want 1 and %.1f%%
#aur agr aap decimal value remove krna chahte hai toh use %d%%
df['Browser'].value_counts().plot(kind='pie' , autopct='%.2f%%')
```

```python
plt.pie(df['Browser'].value_counts(), labels=['Chrome', 'Internet Explorer','Firefox'], autopct = '%d%%')
plt.show()
```

##BAR chart
#it is a chart to visualize categorical features .it shows count vs value analysis for the given features

```python
df['Day'].value_counts()
```

```python
plt.figure(figsize=(8,6))
df['Day'].value_counts().plot(kind='bar')
plt.title('DAYS')
plt.xlabel('WeekDays')
plt.ylabel('Count of the weekdays')
plt.show()
```
