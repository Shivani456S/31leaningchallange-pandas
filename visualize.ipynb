##Area vs. Production
Analysis: Calculate the correlation between chickpea area and production.
Visualization: Scatter Plot
How: Create a scatter plot with chickpea area on the x-axis and production on the y-axis. Add a trend line to indicate the correlation

```python
from scipy import stats
chickpea_by_area = df['CHICKPEA AREA (1000 ha)']
print(chickpea_by_area)
```

```python
chickpea_by_production = df['CHICKPEA PRODUCTION (1000 tons)']
print(chickpea_by_production)
```

```python
correlation_area_by_production = chickpea_by_area.corr(chickpea_by_production)
print(correlation_area_by_production)
```

```python
plt.figure(figsize=(6,5))
plt.scatter(chickpea_by_area, chickpea_by_production, color='lightblue', alpha=0.3, label='Data Points')
slope, intercept, r_value, p_value, std_err = stats.linregress(chickpea_by_area, chickpea_by_production)
plt.plot(chickpea_by_area, slope * chickpea_by_area + intercept, color='blue', label=f'Trend Line (r={correlation_area_by_production:.2f})')
plt.title('Distribution by Correlation in Area and Production',color='blue')
plt.xlabel('Area',color='blue')
plt.ylabel('Production',color='blue')
plt.xticks()
plt.show()
```

##Sorghum Production Patterns
#Analysis: Identify planting patterns for kharif and rabi sorghum.
#Visualization: Heatmap
#How: Create a heatmap showing the area planted for kharif and rabi sorghum across different districts, allowing for visual pattern recognition.

```python
import seaborn as sns
import matplotlib.pyplot as plt
df_filtered = df[['Dist Name', 'KHARIF SORGHUM AREA (1000 ha)' , 'RABI SORGHUM AREA (1000 ha)']]
heatmap_area = df_filtered.set_index('Dist Name').T 
plt.figure(figsize=(6,7))
sns.heatmap(heatmap_area, cmap='YlGnBu', annot=True, fmt='.1f', linewidths=0.5, cbar_kws={'label': 'Area (1000 ha)'})
plt.title('Heatmap Of Planting Patterns by_Rabi & Kharif')
plt.xlabel('Sorghum types')
plt.ylabel('Disticts Name')
plt.tight_layout()
plt.show()
```
