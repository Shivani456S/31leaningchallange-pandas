##QUESTION 3:Show the relationship between the number of vehicles and the average electric range.
Use a scatter plot to illustrate the correlation between vehicle count and average electric range.

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\Users\chaud\OneDrive\Desktop\UNIT2PYTHONLIBRARIES\Electric_Vehicle_Population_Data.csv")
df.head()
```

##QUESTION 4: What is the trend of average electric range yearly?
Plot a line graph to represent the average electric range for each model year.

##QUESTION 6: What percentage of vehicles are eligible for Clean Alternative Fuel Vehicle (CAFV)?
Create a pie chart to show the distribution of CAFV eligibility.

```python
import pandas as pa
import matplotlib.pyplot as plt
df=pa.read_csv(r"C:\Users\chaud\OneDrive\Desktop\UNIT2PYTHONLIBRARIES\Electric_Vehicle_Population_Data.csv")
df.head()
```

```python
#Clean Alternative Fuel Vehicle (CAFV) Eligibility COLUMN NAME ka count nikala 
CAFV_counts = df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].value_counts()
print(CAFV_counts)
```

```python
#ab hum iss column ka percentage nikalenge aur hum ab percentage nikalne ka formula lgayenge
CAFV_percentage = (CAFV_counts / CAFV_counts.sum()) * 100
print(CAFV_percentage)
```

```python
#ab kya krenge humne jo bhi data clean kiya hai uska plot krenge mtlb chart bnayenge
plt.figure(figsize=(8,8))
plt.pie(CAFV_percentage, labels=CAFV_percentage.index, autopct='%1.1f%%', startangle=135, colors=['yellow','pink','green'])
plt.title('CAFV distribution Eligibility')
plt.show()
```

##QUESTION 7: Which city has the highest average base MSRP for electric vehicles?
Use a bar chart to display the average base MSRP by city.

```python
import pandas as pa
import matplotlib.pyplot as plt
df = pa.read_csv(r"C:\Users\chaud\OneDrive\Desktop\UNIT2PYTHONLIBRARIES\Electric_Vehicle_Population_Data.csv")
df.head()
```

```python
#ab kya krenge grouby krenge city aur Base mrsp ka aur mean value nikal lenge 
average_MRSP_BY_CITY = df.groupby('City')['Base MSRP'].mean()
print(average_MRSP_BY_CITY)
```

```python
plt.figure(figsize=(10,6))
average_MRSP_BY_CITY.sort_values(ascending=False).plot(kind='bar', color='yellow')
plt.title('Average Base  MSRP  of Electric Vehicle by City')
plt.xlabel('City')
plt.ylabel('Base MSRP')
plt.show()
```

```python
#upr waale bar chart me hume kuch bhi smjh nahi arha tha toh humne kya kiya accha dikhe bar chart toh humne top starting 20 city leli,head(20)
top_20_city = average_MRSP_BY_CITY.sort_values(ascending=False).head(20)
plt.figure(figsize=(10,6))
top_20_city.plot(kind='bar', color='yellow')
plt.title('Top 20 Average Base  MSRP  of Electric Vehicle by City')
plt.xlabel('City')
plt.ylabel('Base MSRP')
plt.show()
```
